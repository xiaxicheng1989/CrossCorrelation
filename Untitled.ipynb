```python
from scipy import signal
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
matplotlib.rcParams['figure.figsize'] = (16,4)
```

I would like to use cross-correlation to extract the pure inertial signal from two MEMS with uncorrelated noise. 

For the start, Let us create an underlying true sine signal, which we want to extract.

```python
def signal(flist, amplist, offlist):
    t = np.arange(-1,1,0.0001)
    sig = np.zeros(len(t))

    for freq, amp, offset in zip(flist, amplist,offlist):
        #signtemp = [offset + tx if tx<=0 else (offset - tx) for tx in t]
        #signtemp = [offset + tx for tx in t] 
        signtemp = offset +amp*np.sin(2*np.pi*freq*t)
        sig = np.add(sig, signtemp)
    return t, sig
```

```python
freqList = [20,5,40,2,150]
ampList =np.ones(len(freqList))
offList = np.zeros(len(freqList))
time, signal0 = signal(freqList, ampList, offList)
```

```python
time, signal0 = signal(freqList, ampList, offList)
plt.plot(time, signal0)
plt.xlabel('Time')
plt.ylabel('Signal Magnitude')
```

Now we can generate signal. We need to make sure that we have a FFT for this too.

```python
def getFFT(xdata, ydata, dt):
    testfft = np.fft.fft(ydata)/len(ydata)
    freq = np.fft.fftfreq(xdata.shape[-1],dt)
    return freq, testfft

def getFFT_abs(freq, fftdata):
    return freq[0:int(len(freq)/2)], abs(fftdata[0:int(len(fftdata)/2)])
def getPSD(freq, fftdata):
    return freq[0:int(len(freq)/2)], abs(fftdata[0:int(len(fftdata)/2)])**2
```

```python
freq, testfft = getFFT(time, signal0,0.0001)
freq, testpsd = getPSD(freq, testfft)
```

```python
plt.loglog(freq, testpsd)
```

Looks good sofar.  
Let us now generate two identical signal with 2 uncorrelcted noise spcectra.

```python
getSignal = lambda sigma: signal0+np.random.normal(0, sigma, len(time))
```

```python
freqList = [100,350,500,10]
ampList =np.ones(len(freqList))
offList = np.zeros(len(freqList))
time, signal0 = signal(freqList, ampList, offList)
signal1 = getSignal(1)
signal2 = getSignal(1)
plt.plot(time, signal1)
plt.plot(time, signal2)
plt.plot(time, signal0)
plt.legend(['noisy 1', 'noisy 2', 'True signal'])
```

```python
freq, testfft1 = getFFT(time, signal1,0.0001)
freq, testfft2 = getFFT(time, signal2,0.0001)
freq1, testpsd1 = getPSD(freq, testfft1)
freq2, testpsd2 = getPSD(freq, testfft2)
```

```python
plt.loglog(freq2, testpsd2)
plt.loglog(freq1, testpsd1)
plt.legend(['noisy 1', 'noisy 2'])
```

These also make sense. Now let us try the correlation thingy. 

```python
xpsd=abs(np.conj(testfft1)*testfft2)[0:int(len(freq)/2)]**2
```

```python
plt.loglog(freq2, testpsd2)
plt.loglog(freq1, testpsd1)
plt.loglog(freq1, xpsd)
plt.legend(['noisy 1', 'noisy 2', 'X-correlated'])
plt.xlabel('Frequency [Hz]')
plt.ylabel('PSD')
```

This looks promising. Let us now put this into a loop.

```python
sigma1=4
sigma2=2

signal1_0 = getSignal(sigma1)
signal2_0 = getSignal(sigma2)

freq, testfft1 = getFFT(time, signal1_0,0.0001)
freq, testfft2 = getFFT(time, signal2_0,0.0001)
freq1, testpsd1 = getPSD(freq, testfft1)
freq2, testpsd2 = getPSD(freq, testfft2)

xfft=[]
xpsd=[]
for runNr in range(1000):
    freq, indfft1 = getFFT(time, getSignal(sigma1), 0.0001)
    freq, indfft2 = getFFT(time, getSignal(sigma2), 0.0001)
    xfft.append(np.conj(indfft1)*indfft2)
    xpsd.append(abs(np.conj(indfft1)*indfft2)[0:int(len(freq)/2)]**2)
    if runNr ==0:
        xpsd_1runs=np.mean(np.array(xpsd),axis=0)
        xfft_1runs=np.mean(np.array(xfft),axis=0)
    if runNr ==9:
        xpsd_10runs=np.mean(np.array(xpsd),axis=0)
        xfft_10runs=np.mean(np.array(xfft),axis=0)
    if runNr ==19:
        xpsd_20runs=np.mean(np.array(xpsd),axis=0)
        xfft_20runs=np.mean(np.array(xfft),axis=0)
    if runNr ==999:
        xpsd_1000runs=np.mean(np.array(xpsd),axis=0)
        xfft_1000runs=np.mean(np.array(xfft),axis=0)
```

```python
fig, [ax1,ax2,ax3] = plt.subplots(3,1, figsize=(16,12))
ax1.plot(time,signal1_0)
ax1.plot(time,signal2_0)
ax1.plot(time,signal0)
ax1.legend(['Signal 1', 'Signal 2', "Underlying signal"])
ax1.set_xlabel('Time [Hz]')
ax1.set_ylabel('Voltage')

ax2.loglog(freq2, testpsd2)
ax2.loglog(freq1, testpsd1)
#ax2.loglog(freq1, xpsd_1runs)
#ax2.loglog(freq1, xpsd_10runs)
ax2.loglog(freq1, xpsd_1000runs)
ax2.legend(['Signal 1', 'Signal 2', #"1 x xCorr", "10 x xCorr", 
            "100 x xCorr"])
ax2.set_xlabel('Freq [Hz]')
ax2.set_ylabel('PSD')
#ax2.set_ylim([10**(-6),1])


#ax3.plot(time,2000*np.fft.ifft(xfft_10runs).real)
ax3.plot(time,signal0)
ax3.plot(time,40000*np.fft.ifft(xfft_1000runs)+5)
ax3.legend(['Input', 'Reconstructed'])
```
